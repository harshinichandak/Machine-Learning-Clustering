```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns = pd.read_csv(r'D:\Datasets\snsdata.csv')
```

```python
sns.head()
```

```python
sns.isnull().sum()
```

```python
sns.age.value_counts()
```

```python
sns.gender.fillna('F',   inplace = True)
sns.age.fillna(sns.age.mean(),   inplace = True)
```

```python
sns.columns[ sns.dtypes == 'object']
```

```python
sns.gender = sns.gender.replace({'F':1, 'M':0})
```

```python
#sns.info()
```

```python
from sklearn.cluster import KMeans
Kmeans_sns = KMeans(n_clusters = 3)
```

```python
Kmeans_sns.fit(sns)   # model has been created at this step
```

```python
list(Kmeans_sns.labels_)
```

```python
len(list(Kmeans_sns.labels_))
```

```python
Kmeans_sns.cluster_centers_
```

```python
Kmeans_sns.score(sns) # this is the ssd for cluster = 3
```

```python
# now we can build our elbow plot
```

```python
ssd = []
for k in range(1,11):
    Kmeans_sns = KMeans(n_clusters = k)
    Kmeans_sns.fit(sns)
    score = Kmeans_sns.score(sns)
    score = np.round(np.abs(score))
    ssd.append(score)
```

```python
ssd
```

```python
clusters = list(range(1,11))
clusters
```

```python
plt.plot(clusters, ssd, marker = '*')
plt.xlabel('Number of clusters')
plt.ylabel('Sum Square distance')
plt.title('Elbow plot on SNS data')
plt.grid()
```
