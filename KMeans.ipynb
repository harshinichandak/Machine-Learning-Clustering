```python
ssd = []
for k in range(1,11):
    Kmeans_cd = KMeans(n_clusters = k)
    Kmeans_cd.fit(cd)
    score = Kmeans_cd.score(cd)
    score = np.round(np.abs(score))
    ssd.append(score)
```

```python
clusters = list(range(1,11))
clusters
```

```python
plt.plot(clusters, ssd, marker = '*')
plt.xlabel('Number of clusters')
plt.ylabel('Sum Square distance')
plt.title('Elbow plot on SNS data')
plt.grid()
```
